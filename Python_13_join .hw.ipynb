```python
import pandas as pd
ratings = pd.read_csv('ratings.csv')
life_time = ratings.groupby(
    'userId').agg({'timestamp': ['min', 'max', 'count']}
                 ).reset_index()
ratings = ratings.groupby('userId').count().reset_index()
ratings_ = ratings[(ratings['rating'] > 100)]
joined = ratings_.merge(life_time, how='left', on = 'userId')
joined['life_time'] = joined['timestamp', 'max'] - joined['timestamp', 'min']
print('среднее время жизни пользователей, которые выставили более 100 оценок:',
      round(joined['life_time'].mean(),))
```
