```python
import pandas as pd
ratings = pd.read_csv('ratings.csv')
ratings_ = 5.0
movie = ratings[ratings.rating == ratings_].drop_duplicates('movieId', keep='last').agg([max])
print('ID фильма с максимальным количеством оценки "5":', movie['movieId'][0])
```

```python
power = pd.read_csv('power.csv')
period = power[(2005 <= power.year) & (power.year <= 2010)]
latvia = period[period.country == 'Latvia']
latvia_4 = latvia[latvia.category == 4] 
latvia_quantity_4 = latvia_4[latvia_4.quantity >= 0]['quantity'].agg([sum])
latvia_12 = latvia[latvia.category == 12] 
latvia_quantity_12 = latvia_12[latvia_12.quantity >= 0]['quantity'].agg([sum])
latvia_21 = latvia[latvia.category == 21]
latvia_quantity_21 = latvia_21[latvia_21.quantity >= 0]['quantity'].agg([sum])
latvia_quantity = latvia_quantity_4[0] + latvia_quantity_12[0] + latvia_quantity_21[0]
lithuania = period[period.country == 'Lithuania']
lithuania_4 = lithuania[lithuania.category == 4] 
lithuania_quantity_4 = lithuania_4[lithuania_4.quantity >= 0]['quantity'].agg([sum])
lithuania_12 = lithuania[lithuania.category == 12] 
lithuania_quantity_12 = lithuania_12[lithuania_12.quantity >= 0]['quantity'].agg([sum])
lithuania_21 = lithuania[lithuania.category == 21]
lithuania_quantity_21 = lithuania_21[lithuania_21.quantity >= 0]['quantity'].agg([sum])
lithuania_quantity = lithuania_quantity_4[0] + lithuania_quantity_12[0] + lithuania_quantity_21[0]
estonia = period[period.country == 'Estonia']
estonia_4 = estonia[estonia.category == 4] 
estonia_quantity_4 = estonia_4[estonia_4.quantity >= 0]['quantity'].agg([sum])
estonia_12 = estonia[estonia.category == 12] 
estonia_quantity_12 = estonia_12[estonia_12.quantity >= 0]['quantity'].agg([sum])
estonia_21 = estonia[estonia.category == 21]
estonia_quantity_21 = estonia_21[estonia_21.quantity >= 0]['quantity'].agg([sum])
estonia_quantity = estonia_quantity_4[0] + estonia_quantity_12[0] + estonia_quantity_21[0]
quantity = latvia_quantity + lithuania_quantity + estonia_quantity
print('Cуммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 годы составило:', quantity)
```
