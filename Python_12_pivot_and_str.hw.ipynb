```python
import pandas as pd
log = pd.read_csv('visit_log.csv', sep=';')
a = log.loc[
    (
        (log.traffic_source == 'yandex') | 
         (log.traffic_source == 'google')
    ), 'source_type'
] = 'organic'
log.loc[
    (
        (log.region == 'Russia') & 
        ((log.traffic_source == 'paid') |
        (log.traffic_source == 'email'))
        ), 'source_type'
] = 'ad'
log.loc[
    (
        (log.region != 'Russia') & 
        ((log.traffic_source == 'paid') |
        (log.traffic_source == 'email'))
        ), 'source_type'
] = 'other'
log
```

```python
# не работает Mystem()
import re
data = pd.DataFrame({
    'keyword': ['курс гривны к рублю', 'доллары в рубли', '100 долларов в рублях', 'курс рубля'],
    'shows': [125076, 114173, 97534, 53546],
})
def make_lemmas(keyword):   
    regex = r'\b[а-я]{4}'
    lemmas = re.findall(regex, keyword, re.I)
    return ' '.join(lemmas)
data['lemmas'] = data['keyword'].apply(make_lemmas)
data
```
