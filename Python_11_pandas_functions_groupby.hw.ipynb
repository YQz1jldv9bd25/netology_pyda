```python
import pandas as pd
ratings = pd.read_csv('ratings.csv')
ratings
rating = ratings['rating']
def class_movies(rating):     
    if 0.0 < rating <= 2.0:
        x = 'низкий рейтинг'
    if 2.0 < rating <= 4.0:
        x = 'средний рейтинг'
    if rating == 4.5 or rating == 5.0:
        x = 'высокий рейтинг'    
    return x
ratings['class'] = ratings['rating'].apply(class_movies)
ratings
```

```python
keywords = pd.read_csv('keywords.csv')
keywords
keyword = keywords['keyword']
def class_cities(keyword):     
    if 'москва' or 'тула' or 'ярославль' in keyword:
        x = 'Центр'
    if 'петербург' or 'псков' or 'мурманск' in keyword:
        x = 'Северо-Запад'
    if 'владивосток' or 'сахалин' or 'хабаровск' in keyword:
        x = 'Дальний Восток'
    if'москва' or 'тула' or 'ярославль' or'петербург' or 'псков' or 'мурманск' or 'владивосток' or 'сахалин' or 'хабаровск' not in keyword:
        x = 'undefined'
    return x
keywords['region'] = keywords['keyword'].apply(class_cities)
keywords
```

```python
movies = pd.read_csv('movies.csv')
def production_year(title):
    for year in range(1950, 2011):
        if str(year) in title:
            return year
    return 1900
movies['year'] = movies['title'].apply(production_year)
years = movies[['movieId','year']]
joined_ratings = ratings.merge(years, on='movieId', how='left') 
count_year = joined_ratings.groupby('year').mean()
count_year.sort_values(by=['rating'], ascending=False).reset_index()
```
